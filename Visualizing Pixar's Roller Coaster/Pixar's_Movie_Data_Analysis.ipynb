# Data Exploration

```python
# Import the libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Loading the dataset.

pixar_movies = pd.read_csv("PixarMovies.csv")
pixar_movies.shape
pixar_movies.head(pixar_movies.shape[0])
```

```python
# Display datatypes of each column.
pixar_movies.dtypes 
```

```python
# A summary table
pixar_movies.describe()
```

# Data Cleaning

```python
pixar_movies = pixar_movies.dropna() #Remove NA rows from the dataset.
pixar_movies
```

```python
# Strip the % sign from the column to do calculations and convert them to float.
pixar_movies["Domestic %"] = pixar_movies["Domestic %"].str.rstrip("%").astype("float")
pixar_movies["International %"] = pixar_movies["International %"].str.rstrip("%").astype("float")

#Feature engineering.
pixar_movies["IMDB Score"] = pixar_movies["IMDB Score"] * 10
pixar_movies
```

 # Data Visualization 

```python
# Line Plot
pixar_movies.plot(x = "Movie",y = ["RT Score","IMDB Score","Metacritic Score"],figsize = (10,6))
plt.show()
```

```python
#Boxplots

critics_reviews = pixar_movies[["RT Score","Metacritic Score","IMDB Score"]]
critics_reviews.plot(kind = "box", figsize = (9,5))
```

From the above Boxplot, it can be easily inferred that Totten Tomatoes always have given very good ratings for Pixar movies. The Metacritic ratings are widely spread out from higher 50s to mid 90s. And IMBD ratings are relatively of lower numbers compared to other two.

```python
#Stacked Bar Plot
revenue_proportions = pixar_movies[["Movie","Domestic %","International %"]]
revenue_proportions.plot(x= 'Movie',kind = 'bar', stacked = True)
```

From the above image, we can infer that te first few movie releases from Pixar had 50% whereas the last few releases are somewhere around 40%.

```python
# Grouped Bar Plot
pixar_movies[["Oscars Nominated","Oscars Won"]].plot(kind = 'bar', stacked = True)
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```

Getting more nominations increases the chance of landing an oscar by a pixar movie, except getting six nominations.
